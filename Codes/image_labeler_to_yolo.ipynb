```python
from PIL import Image
import os, sys
```

```python
def convert(box):
    dw = 1./640
    dh = 1./640
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return x,y,w,h

# reading a file
notation_file_path = r"Dataset\PCBData\group00041\00041_not//"
notation_file_dirs = os.listdir(notation_file_path)

saving_folder = r"Dataset\PCBData\group00041\labels//"
saving_folder_dirs = os.listdir(saving_folder)

image_file_path = r"Dataset\PCBData\group00041\00041//"
image_file_dirs = os.listdir(image_file_path)

def annotate(argument):
    switcher = {
        0: "background ",
        1: "open",
        2: "short",
        3: "mousebite",
        4: "spur",
        5: "copper",
        6: "pinHole"
    }
    return switcher.get(argument, "nothing")


for imag in image_file_dirs:
    if os.path.isfile(image_file_path+imag):
        im = Image.open(image_file_path+imag)
        f,e = os.path.splitext(image_file_path+imag)
        if f.endswith("test"):
            fk = f.split("//")[-1]
            fn = f.split("//")[-1].split("_")[0]
            text_file = open(notation_file_path+fn+".txt")
            text_lines = text_file.readlines()
            count = 0
            arb_list = [""]
            for line in text_lines:
                nf,ne = os.path.splitext(saving_folder)
                x1,y1,x2,y2,anno = [int(s) for s in line.strip().split()]
                string_line = [str(i) for i in [x1, y1, x2, y2, anno]]
                our_line = [""] + string_line

                b = (x1, x2, y1, y2)
                bb = convert(b)
                xc,yc,wy,hy = bb

                our_line = " ".join(string_line)
                # arb_list.append()

                # cropim = im.crop((x1,y1,x2,y2))
                count = count+1
                # print(xc,yc,wy,hy, anno, count)
                yolo_string_line = [str(i) for i in [anno, xc, yc, wy, hy]]
                yolo_our_line = " ".join(yolo_string_line)
                # print(yolo_our_line)

                file_name_txt = nf+fk+".txt"
                print(file_name_txt)

                with open(file_name_txt,'a') as tf:
                    tf.writelines(' '.join(yolo_string_line))
                    tf.writelines("\n")
                # print(nf+str(anno)+"_"+f+"_"+str(annotate(anno))+"_"+str(count))
                # cropim.save(nf+str(anno)+"_"+f+"_"+str(annotate(anno))+"_"+str(count)+".jpeg", "jpeg")
```
